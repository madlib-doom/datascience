```python
import seaborn as sns
import matplotlib.pyplot as plt

sns.swarmplot(x="Gender",y="Age",data=data,palette="spring",hue="CustomerID")
```

```python
sns.barplot(x="Age",y="Annual Income",data=data,palette="autumn",hue="Gender")
```

```python
import matplotlib.dates as mdates

plt.gca().xaxis.set_major_formatter(mdates.DateFormatter("%b"))
plt.gca().xaxis.set_major_locator(mdates.MonthLocator())

sns.lineplot(x="Gender",y="Annual Income",data=data,palette="spring")
```
